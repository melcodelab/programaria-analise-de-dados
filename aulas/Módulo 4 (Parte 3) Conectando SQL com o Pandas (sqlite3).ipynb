```python
pd.read_sql("""SELECT mb.Estado, avg(ms.renda) as media_renda_por_estado
            FROM Municipios_Brasileiros mb
            INNER JOIN Municipio_Status ms
            ON ms.municipio_id = mb.Municipio_ID
            GROUP BY mb.Estado""", con=conexao)
```

```python
lista_estados = list(dados['UF ONDE MORA'].unique())

query = """SELECT mb.Estado, avg(ms.renda) as media_renda_por_estado
            FROM Municipios_Brasileiros mb
            INNER JOIN Municipio_Status ms
            ON ms.municipio_id = mb.Municipio_ID
            WHERE mb.Estado IN ({})
            GROUP BY mb.Estado;""".format(','.join(['?' for _ in lista_estados]))

estado_renda = pd.read_sql(query, con=conexao, params=lista_estados)
```

```python
dados.rename(columns={'UF ONDE MORA': 'Estado'}, inplace=True)
```

```python
dados.columns
```

```python
dados = dados.merge(estado_renda, on="Estado", how='left')
```
